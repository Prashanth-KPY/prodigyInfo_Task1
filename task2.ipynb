```python
df=df.fillna(method="ffill")
df.head()
```

```python
print(df.isnull().sum())
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
df
```

```python
survived_counts=df['Survived'].value_counts()
plt.bar(survived_counts.index,survived_counts.values)
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Count')
plt.show()
```

```python
survivals=df.groupby('Sex')['Survived'].mean()
survivals.plot(kind='bar',color='green')
plt.xlabel('Sex')
plt.ylabel('Survival Rat')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
plt.hist(df['Age'],bins=25,color='yellow',edgecolor='red')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age distribution of Passengers')
plt.show()
```

```python
df.drop(['Cabin'],axis=1,inplace=True)
df.head()
```

```python
x=df['Pclass'].value_counts()
plt.pie(x,labels=x.index,startangle=99,colors=['red','yellow','pink'])
plt.title('Passenger Class Distribution')
plt.show()
```

```python
import seaborn as sn
x=df.corr()
sn.heatmap(x,annot=True,cmap='BuPu')
```

```python
plt.fill_between(np.linspace(0,10,100),np.sin(np.linspace(0,10,100)),np.cos(np.linspace(0,10,100)),color='gold')
plt.title('Area Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
plt.scatter(np.random.rand(32),np.random.rand(32),color='red')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
a=df.groupby('Age')['Fare'].sum().cumsum()
plt.plot(a.index,a.values,marker='o',linestyle='-')
plt.title('Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.boxplot([df[df['Pclass']==1]['Age'],df[df['Pclass']==2]['Age'],df[df['Pclass']==3]['Age']],
            labels=['1st Class','2nd Class','3rd Class'],vert=False,patch_artist=True,
            boxprops=dict(facecolor='yellow'))
plt.title('Age distribution by Passenger class')
plt.xlabel('Age')
plt.ylabel('Passenger Class')
plt.show()
```
